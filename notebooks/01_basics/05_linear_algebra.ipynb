# Eigenvectors and Eigenvalues

```python
import sympy
```

```python
M = matrix([
    [3, -2, 4, -2],
    [5, 3, -3, -2],
    [5, -2, 2, -2],
    [5, -2, -3, 3]
])

M
```

Using sympy:

```python
sympy.Matrix(M).eigenvals()
```

```python
sympy.Matrix(M).eigenvects()
```

Using Sage:

```python
M.eigenvalues()
```

```python
M.eigenvectors_right()
```

```python
M.eigenmatrix_right()
```
