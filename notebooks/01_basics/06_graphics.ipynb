# Overlay

Taylor expansion of $sin(x)$ at $x=0$ at various degrees:

```python
def p(x, n):
    return taylor(sin(x), x, 0, n)
```

```python
xmax = 15
n = 15
```

```python
g = plot(sin(x), x, -xmax, xmax)
g
```

```python
for d in range(n):
    g += plot(p(x, 2 * d + 1), x, -xmax, xmax, \
              color = (1.7*d/(2*n), 1.5*d/(2*n), 1-3*d/(4*n))
             )
```

```python
g.show(ymin=-2, ymax=2)
```

```python
a.show()
```

# Displaying Solutions Of Differential Equations

```python
x = var('x')
y = function('y')
```

```python
DE = x * diff(y(x), x) == 2*y(x) + x^3
DE
```

```python
desolve(DE, [y(x), x])
```

```python
sol = []

for i in srange(-2, 2, 0.2):
    sol.append(desolve(DE, [y(x), x], ics=[1, i]))
    sol.append(desolve(DE, [y(x), x], ics=[-1, i]))
sol
```

```python
g = plot(sol, x, -2, 2)
g.show()
```
