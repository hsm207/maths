# One-to-One and Onto functions

Source: https://www.cs.colorado.edu/~srirams/courses/csci2824-spr14/functionTypes-18.html

Assume:

$$f: \mathbb{N} \rightarrow \mathbb{N}$$

**Sample Question 1**

$$f(n) = n^2$$ is one-to-one but not onto:

```python
f(n) = n^2
f
```

```python
plot(f(n), n, 1, 100)
```

Notice that every $n$ is assigned a unique $f(n)$.

Not onto:

```python
B = [f(n) for n in range(1, 11)]
B
```

There is an element in B e.g. 3 that can't be mapped back to A

**Sample Question 2**

```python
f(x) = 1/x
g(x) = log(x)
```
