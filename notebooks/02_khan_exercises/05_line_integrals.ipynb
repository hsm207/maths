# Line Integrals of Scalar Functions

From [here](https://www.khanacademy.org/math/multivariable-calculus/integrating-multivariable-functions/line-integrals/e/computing-line-integrals-of-scalar-functions?modal=1)

## Question 1

```python
# order matters! define f(x, y) then x(t), y(t)
t = var('t')

f(x,y) = x^2

x(t) = 6 * t
y(t) = t
```

```python
C = Curve((x(t)._sympy_(), y(t)._sympy_()), (t._sympy_(), -1, 1))
```

```python
line_integrate(f(x,y)._sympy_(), C, [sympy.symbols('x'), sympy.symbols('y')])
```

```python
t = var('t')

f(x,y) = x - y

x(t) = 2 * cos(t)
y(t) = 2 * sin(t)


C = Curve((x(t)._sympy_(), y(t)._sympy_()), (t._sympy_(), 0, pi))

line_integrate(f(x,y)._sympy_(), C, [sympy.symbols('x'), sympy.symbols('y')])
```

## Line Integrals in Vector Fields

From [here](https://www.khanacademy.org/math/multivariable-calculus/integrating-multivariable-functions/line-integrals-vectors/e/computing-line-integrals-in-vector-fields?modal=1).

### Question 1



```python
import sympy
from sympy.vector import CoordSys3D, ParametricRegion, vector_integrate

C = CoordSys3D('C')

t = var('t')._sympy_()
x = var('x')._sympy_()
y = var('y')._sympy_()

region = ParametricRegion((cos(t), sin(t)), (t, 0, pi/2))
field = (C.x/C.y) * C.i + (C.y/C.x) * C.j

vector_integrate(field, region)
```

### Question 2

```python
import sympy
from sympy.vector import CoordSys3D, ParametricRegion, vector_integrate

C = CoordSys3D('C')

t = var('t')._sympy_()
x = var('x')._sympy_()
y = var('y')._sympy_()

region = ParametricRegion((t^2, 3*t), (t, 0, 2))
field = sympy.sqrt(C.x) * C.i + (2 * C.y) * C.j

vector_integrate(field, region)
```

### Question 3

```python
import sympy
from sympy.vector import CoordSys3D, ParametricRegion, vector_integrate

C = CoordSys3D('C')

t = var('t')._sympy_()
x = var('x')._sympy_()
y = var('y')._sympy_()

region = ParametricRegion((t, 2*t), (t, 0, 1))
field = (C.y^3 - 4*C.y)*C.i + (C.x^3 - 4*C.x)*C.j

vector_integrate(field, region)
```
