```python
x, y = var('x'), function('y')(x)

eqn1 = diff(y, x) == x - y
eqn1
```

```python
soln1 = desolve(eqn1, y)
soln1
```

```python
C = soln1.variables()[0]
C
```

```python
C_val = solve(soln1.substitute(x == 3) == 1, C)[0].rhs()
C_val
```
