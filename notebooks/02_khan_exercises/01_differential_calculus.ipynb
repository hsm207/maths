# Question 15



The goal is to find the derivative of $c(t)$ wrt $t$:

```python
t = var('t')
b = var('b')

a = function('a')(t)
c = function('c')(t)
```

```python
eqn1 = c^2 == a^2 + b^2
eqn1
```

```python
soln = solve(eqn1, c)
soln
```

Distance must be positive, so we take the positive solution:

```python
soln[1].rhs()\
    .diff(t)\
    .substitute(b == 30, a == 16, diff(a, t) == 16)\
    .n(digits=4)
```
